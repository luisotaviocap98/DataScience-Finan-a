```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Lendo a base de dados
Base_Dados = pd.read_csv('../Pandas/StudentsPerformance.csv')
```

```python
Base_Dados
```

```python
plt.figure(figsize=(8,5)) #largura altura
sns.scatterplot(data=Base_Dados, x='math score', y='writing score', hue='gender', style='race/ethnicity')
```

```python
m = {'female':'o', 'male':'X'}
sns.scatterplot(data=Base_Dados, x='math score', y='writing score', hue='gender', style='gender', markers=m)
```

```python
sns.boxplot(data=Base_Dados, y='math score', x='race/ethnicity')
plt.yticks([i for i in range(0,101,10)])
plt.grid(axis='y',alpha=0.5)
```

```python
sns.boxplot(data=Base_Dados, y='math score', x='race/ethnicity', hue='gender')
```
