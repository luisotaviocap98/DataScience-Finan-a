```python
def troca(linha):
    return linha.split('R$')[-1].replace('.','').replace(',','.').strip()
```

```python
novo['NovaRenda'] = novo['Renda'].apply(troca)
```

```python
ordem = sorted(novo['NovaRenda'].unique(), reverse=True)
```
