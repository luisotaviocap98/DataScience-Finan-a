```python
btc = yf.Ticker('BTC-USD').history(period='max')[['Open', 'Close', 'High', 'Low', 'Volume']]
```

```python
temp = btc['2021-1-1':]
```

```python
f = go.Figure(data=[
    go.Candlestick(
        x=temp.index,
        open= temp['Open'],
        high=temp['High'],
        low=temp['Low'],
        close=temp['Close'],
        showlegend=False
    )
])

lay = go.Layout(
    yaxis_tickformat='$',
    xaxis=dict(rangeslider=dict(visible=False))
)

f.update_layout(lay)

f.show()
```
